```
# -*- coding: utf8 -*-
# http://itnovella.ru/itnovella/2013/10/21/matplotlib.html
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(-10,10.01,0.01) #Массив значений аргумента
plt.plot(x,x**2) #Построение графика
plt.xlabel(r'$x$') #Метка по оси x в формате TeX
plt.ylabel(r'$f(x)$') #Метка по оси y в формате TeX
plt.title(r'$y=x^2$') #Заголовок в формате TeX
plt.grid(True) #Сетка
plt.show() #Показать график
```

```
import numpy as np
import matplotlib.pyplot as plt

def f1(x):
  return -4*(x**2)+3*x+2

x=np.arange(-1.25,2.01,0.01) #Массив значений аргумента
plt.plot(x,f1(x)) #Построение графика
#plt.grid(True) #Сетка
plt.show() #Показать график
```
